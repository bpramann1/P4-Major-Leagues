```python
#So a lot of these columns seem to be normally distributed! I want them to be uniformally distributed
#So I will created a couple of bins and then take all the data that fits a certain bin and pull a few records from it
#I want to have as much data as possible so the number of records I will pull is the number of records in the smallest bin
```

```python
numberOfBins = 10
currentColumnName = "elo1_pre"
colLow = mlbData[currentColumnName].min()
colHigh = mlbData[currentColumnName].max()
delta = (colHigh-colLow)/(numberOfBins - 1)
counts = []
for i in range(0, numberOfBins-1):
    counts.append((mlbData.query(currentColumnName + ' >= ' + str(colLow + (i*delta)) + ' and ' + currentColumnName + ' <= ' + str(colLow + ((i+1)*delta)))[currentColumnName]).count())

numberOfElementsPerBin = min(counts)
semiUniformedSample = []
for i in range(0, numberOfBins-1):
    semiUniformedSample.append(mlbData.query(currentColumnName + ' >= ' + str(colLow + (i*delta)) + ' and ' + currentColumnName + ' <= ' + str(colLow + ((i+1)*delta))).sample(n=numberOfElementsPerBin, random_state=1))

uData = pd.DataFrame([], columns = mlbData.columns) 

recordIndex = 0
for sampleBin in semiUniformedSample:
    for i in range(0, numberOfElementsPerBin-1):
        uData.loc[recordIndex] = sampleBin.iloc[i]
        recordIndex = recordIndex + 1
        
uData
```

```python
ax1 = uData.plot.scatter(x='elo1_pre',
                      y='score1',
                      c='DarkBlue')
```

```python
X = uData["elo1_pre"]
y = uData["score1"]

sns.regplot(x='elo1_pre', y='score1', data=uData)
model = sm.regression.linear_model.OLS(y, sm.add_constant(X))
fittedModel = model.fit()
p = fittedModel.params
fittedModel.summary()
#ax = uData.plot(x='elo1_pre', y='score1', kind='scatter')
#ax.plot(X, p.const + p.elo1_pre * X)
```
