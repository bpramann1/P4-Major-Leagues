```python
tagsData.query('movieId == "' + str(260) + '"')['tag']
```

```python
associatedTags = []
for movieId in movieIds:
    associatedTags.append(list(tagsData.query('movieId == "' + str(movieId) + '"')['tag']))
```

```python
#Now that we have an array of arrays of assoicated tags, we can go to a dataframe with two columns( associatedTag1, associatedTag2)
```

```python
#After this point, we have basically the same situation as https://www.kaggle.com/ankur310794/network-analysis-of-hobbies-interests
```

```python
#We can keep in step with that solution until we have a couple of categories to throw a tag into
```

```python
#Then we can go through all the movies and throw them into categories based on the category that has the most tags in common with the movie
```

```python
#Then we have all the movies categorized
```

```python
#Then for we average all the ratings for each movie
```

```python
#Then we line up all the movies from each category from highest rating to lowest
```

```python
#Then we suggest the highest rated movie from the category that the current movie is in
```

```python
#This movie matching algorithm will have no randomness and will not take into account whether or not you have watched the movie before
```

```python
type(associatedTags[0])
```

```python
associations = pd.DataFrame([], columns = ['aTag1', 'aTag2']) 
```

```python
rowIndexA = 0
for aTagList in associatedTags:
    for n in range(0, len(aTagList)-1):
        if((n+1) <= (len(aTagList)-1)):
            for i in range(n+1, len(aTagList)-1):
                associations.loc[rowIndexA] = [aTagList[n],aTagList[i]]
                rowIndexA = rowIndexA + 1
```

```python
#As you can see, the computation above is O(m * n^2) which is crazy computationally inefficent but fortunately n is usually small
```
